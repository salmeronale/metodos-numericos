```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def f(x):
  return math.e**(x/2) * math.sin(x/3)

def g(x):
  return (1/3)*x

def h(x):
  return (1/3)*x + (1/6)*(x**2)

def e(x):
  return (1/3)*x + (1/6)*(x**2) + (23/648)*(x**3)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,200)
```

```python
# Graficar funciones.
plt.figure(figsize=(10,10))
y = [f(i) for i in x]  # Evaluar f(x) para cada x
plt.plot(x,y, label="f(x)")
plt.plot(x,g(x), label="P1(x)")
plt.plot(x,h(x), label="P2(x)")
plt.plot(x,e(x), label="P3(x)")
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title(r'Polinomio de Maclaurin')
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid(color="gray")
plt.legend()
plt.show()
```
