# Gráfica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Función cuadrática.
def f(x):
    return x - 3**(-x)

#Función equis
def g(x):
  return x

#función 3
def h(x):
  return -3**(-x)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(10,10))
plt.plot(x, g(x))
plt.plot(x, h(x))
# al momento de sumar los valores de g(x) y h(x)da como resultado f(x)
plt.plot(x,f(x))
plt.axhline(0, color="purple")
plt.axvline(0, color="purple")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title('MI PRIMERA GRAFICA $f(x)= g(x)+ h(x)$')
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid(color="gray")
plt.show()
```
