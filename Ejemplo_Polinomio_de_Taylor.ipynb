```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
  return (x+1)**(1/2)

def g(x):
  return 1 + (1/2)*x

def h(x):
  return 1 + (1/2)*x - (1/8)*(x**2)

def e(x):
  return 1 + (1/2)*x - (1/8)*(x**2) + (15/16)*(x**3)
```

```python
# Valores del eje X que toma el gráfico.
x=np.linspace(-3, 10,100)
```

```python
# Graficar funciones.
plt.figure(figsize=(10,10))
plt.plot(x,f(x), label="$\sqrt{x+1}$")
plt.plot(x,g(x), label="$1 + (1/2)x$")
plt.plot(x,h(x), label="$1 + (1/2)x - (1/8)x2$")
plt.plot(x,e(x), label="$1 + (1/2)x - (1/8)x2 + (15/16)x3$")
plt.axhline(0, color="purple")
plt.axvline(0, color="purple")
plt.xlim(-3, 10)
plt.ylim(-2, 5)
plt.title(r'Polinomio de Taylor $f(x)=\sqrt{x+1}$')
plt.xlabel("EJE X")
plt.ylabel("EJE Y")
#plt.savefig("output.png") # Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.grid(color="gray")
plt.legend()
plt.show()
```
